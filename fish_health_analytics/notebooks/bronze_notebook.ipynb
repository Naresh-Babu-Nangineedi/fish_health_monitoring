```python
from pyspark.sql.functions import col, current_timestamp, input_file_name
from pyspark.sql.types import TimestampType, DoubleType, StringType
```

```python
file_path = "Files/Data_Model_IoTMLCQ_2024.csv"

df_bronze = (spark.read.option("header", True).option("inferSchema", True).csv(file_path))
```

```python
# Adding metadata columns
df_bronze = (
    df_bronze
        .withColumn("source_file", input_file_name())
        .withColumn("ingestion_timestamp", current_timestamp())
)
```

```python
df_bronze.write.mode("overwrite").saveAsTable("bronze_iot_tilapia_table")
```
